```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
A = 1  # Amplitude
a = 1  # Length scale

# Define the wave function
def psi(x):
    return A * np.sin(2 * np.pi * x / a)

# Generate x values
x_values = np.linspace(-a/2, a/2, 1000)

# Plot the wave function
plt.figure(figsize=(8, 6))
plt.plot(x_values, psi(x_values), label=r'$\psi(x)$', color='b')
plt.title('Space Dependence of Wave Function')
plt.xlabel('x')
plt.ylabel(r'$\psi(x)$')
plt.grid(True)
plt.legend()
plt.show()
```
